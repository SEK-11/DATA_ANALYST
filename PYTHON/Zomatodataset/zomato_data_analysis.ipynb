```python
df_country=pd.read_excel('/content/sample_data/Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.dtypes
```

```python
final_df.value_counts().index
```

```python
#pie chart -Top 3 countries that uses zomato
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation max records are form india

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Count'})
```

```python
ratings
```

Observation

1. rating between 4.5 to 4.9-->excellent

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Count",data=ratings)
```

```python

sns.barplot(x="Aggregate rating",y="Count",hue="Rating color",data=ratings,palette=['black','red','orange','yellow','green','green'])
```

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
